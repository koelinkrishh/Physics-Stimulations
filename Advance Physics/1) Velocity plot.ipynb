```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

Q) Plot velocity and acceleration v/s time for t->[1,100]sec 
for displacement r = 1-2t+t<sup>2<sup/>

```python
Time = np.arange(1,11,0.3) # time
Displacement = np.zeros(Time.shape) # displacement

# Mapping in numpy:
t = sp.Symbol('t') # Defining variable

x = lambda t: t -4*t**2 +6
Displacement = x(Time)
print(Displacement)
```

To calculate velocity and acceleration, <br>
we need to derivate x to get Velocity and Acceleration as function of time <br>
We use sympy to calculate velocity and acceleration

```python
# Converting into Sympy expressions:
Disp = sp.sympify(x(t))
print(Disp)

# Differentiating Displacement:
Vel = sp.diff(Disp,t)
Acc = sp.diff(Vel,t)
print(Vel,' & ',Acc)

# Convert the differentiated expression in callable form:
v = sp.lambdify(t,Vel,'numpy')
a = sp.lambdify(t,Acc,'numpy')
```

```python
# Calculating velocity and acceleration for given range of time:
Velocity = v(Time)
Acceleration = a(Time)

if type(Velocity) == int:
    Velocity = np.full(np.shape(Displacement),Velocity)
if type(Acceleration) == int:
    Acceleration = np.full(np.shape(Displacement),Acceleration)
```

Plotting:

```python
# x v/s t
plt.plot(Time,Displacement,color='blue')
plt.title('Displacement v/s Time')
plt.xlabel('Time [seconds]')
plt.ylabel('Displacement [meters]')
plt.show()
```

```python
# v v/s t
plt.plot(Time,Velocity,color='red')
plt.title('Velocity v/s Time')
plt.xlabel('Time [seconds]')
plt.ylabel('Velocity [meter/second]')
plt.show()
```

```python
# a v/s t
plt.plot(Time,Acceleration,linewidth='5',color='magenta')
plt.title('Accelation v/s Time')
plt.xlabel('Time [seconds]')
plt.ylabel('Acceleration [m/s^2]')
plt.show()
```

```python
# All plots in one frame

plt.plot(Time,Displacement,color='blue',linestyle='-',marker='o',label='Displacement')
plt.plot(Time,Velocity,color='red',label='Velocity')
plt.plot(Time,Acceleration,'-m',label='Acceleration')
plt.title('Motion of object')
plt.xlabel('Time [seconds]')
plt.ylabel('Movement')
plt.legend()
plt.show()
```
